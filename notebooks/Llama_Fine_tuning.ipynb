```python
# Importar dependencias para cargar el modelo entrenado
from transformers import AutoModelForCausalLM, AutoTokenizer
from torch.nn import DataParallel
```

```python
# Se cagar el modelo entrenado
tokenizer = AutoTokenizer.from_pretrained('/content/llamaUAO3')
model = AutoModelForCausalLM.from_pretrained('/content/llamaUAO3')
```

```python
# promp
input_context = '''
###Human:
tell me about pokemon

###Assistant:
'''
```

```python
input_ids = tokenizer.encode(input_context, return_tensors="pt")
output = model.generate(input_ids, max_length=85, temperature=0.3, num_return_sequences=1)
```

```python
# texto generado
generated_text = tokenizer.decode(output[0], skip_special_tokenz=True)
print(generated_text)
```
